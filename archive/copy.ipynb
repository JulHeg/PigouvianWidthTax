```python
import conn_tcp
```

```python
conn = conn_tcp.TCPSocket
```

```python
model = conn(address="192.168.1.13", port=5000, conn_type='zmq')
```

```python
model.open()
```

```python
phase_img, intensity_img, phase_meta, intensity_meta = model.get_image()
import matplotlib.pyplot as plt
import numpy as np
phase = np.nan_to_num(phase_img)
phase = np.clip(phase, 0, 1)
plt.imshow(phase)
plt.colorbar()
plt.show()
```

```python
model.close()
```
